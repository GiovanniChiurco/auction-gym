```python
import pandas as pd
import matplotlib.pyplot as plt
```

## Load data

```python
alpha_0_5 = pd.read_csv('../results/FP_DM_TS_k_LinUCB/agent_stats_0.csv')
alpha_1 = pd.read_csv('../results/FP_DM_TS_k_LinUCB_alpha_1/agent_stats_0.csv')
alpha_1_5 = pd.read_csv('../results/FP_DM_TS_k_LinUCB_alpha_1_5/agent_stats_0.csv')
```

```python
print(f"Number of clicks with alpha = 0.5: {alpha_0_5['Num Clicks'].sum()}")
print(f"Number of clicks with alpha = 1: {alpha_1['Num Clicks'].sum()}")
print(f"Number of clicks with alpha = 1.5: {alpha_1_5['Num Clicks'].sum()}")
```

## Number of clicks per iteration

```python
grouped_0_5 = alpha_0_5.groupby('Iteration').sum()
grouped_1 = alpha_1.groupby('Iteration').sum()
grouped_1_5 = alpha_1_5.groupby('Iteration').sum()
```

```python
plt.plot(grouped_0_5['Num Clicks'], label='alpha = 0.5')
```

```python
plt.plot(grouped_1['Num Clicks'], label='alpha = 1')
```

```python
plt.plot(grouped_1_5['Num Clicks'], label='alpha = 1.5')
plt.show()
```

## Publishers with highest number of clicks

```python
alpha_0_5.groupby('Publisher').sum().sort_values('Num Clicks', ascending=False)
```

```python
alpha_1.groupby('Publisher').sum().sort_values('Num Clicks', ascending=False)
```

```python
alpha_1_5.groupby('Publisher').sum().sort_values('Num Clicks', ascending=False)
```
